```python
#import csv with cleaned marathon data
df_marathondata = pd.read_csv(r'C:\Users\jonas\Desktop\Julia\GitLab\marathons\raw_data\marathon_data_cleaned_summary.csv')
```

```python
#check if marathon dataframe has been imported correctly
df_marathondata
```

```python
df_marathondata.rename(columns={"Platz": "Rank"},inplace=True)
df_marathondata.rename(columns={"Pl.AK": "Rank age group"},inplace=True)
df_marathondata.rename(columns={"Startnr.": "Starting no."},inplace=True)
df_marathondata.rename(columns={"Verein/Team": "Team"},inplace=True)
df_marathondata.rename(columns={"AK": "Age group"},inplace=True)
df_marathondata
```

```python
#group air quality data as we only want the mean for each year
df_airquality_grouped = df_airquality.groupby(["Year"]).agg({"Feinstaub": "mean", "Ozon": "mean", "Stickstoff": "mean"})
df_airquality_grouped
```

```python
#rename the columns in the data frame to consistent English naming

df_airquality_grouped.rename(columns={"Feinstaub": "Particulare matter in in µg/m³ mean"},inplace=True)
df_airquality_grouped.rename(columns={"Ozon": "Ozone in µg/m³"},inplace=True)
df_airquality_grouped.rename(columns={"Stickstoff": "Nitrogene dioxide in µg/m³ mean"},inplace=True)
df_airquality_grouped
```

```python
#group weather  data as we only want the mean for each year + max for temp
df_weatherdata_grouped = df_weatherdata.groupby('Year').agg({'Temp_Celcius': ['mean', 'max'],'Wind_kmperhour': ['mean'], 'Luftfeuchte': ['mean'], 'Barometer_hPa': ['mean']})
df_weatherdata_grouped
```

```python
df_weatherdata_grouped.rename(columns={"Luftfeuchte": "Humidity"},inplace=True)
df_weatherdata_grouped
```

```python
#merge air and weather data to one data frame
df_merged_airandweather = pd.merge(left=df_airquality_grouped, right=df_weatherdata_grouped, how='left', on='Year')
df_merged_airandweather
```

```python
#merge the new air and weather data frame with the marathon dataframe to have one data frame with all information
df_merged_marathon_air_weather = pd.merge(left=df_marathondata, right=df_merged_airandweather, how='left', on='Year')
df_merged_marathon_air_weather
```

```python
#check the datatypes of the new dataframe
df_merged_marathon_air_weather.dtypes
```

```python
#recognized that AK (=Altersklasse/age group) is an object and categorical feature and this cannot be used in correlation matrix. Thus, using one hot encoding for categorical feature "Altersklasse"
one_hot_encoded = pd.get_dummies(df_merged_marathon_air_weather['AK'], dtype=int)
```

```python
one_hot_encoded
```

```python
#merging the one hot encoding outcome to the merged dataframe, result is a new dataframe df_merged_marathon_air_weather_with encoded
df_merged_marathon_air_weather_with_encoded = pd.concat([df_merged_marathon_air_weather,one_hot_encoded], join='inner', axis=1)
```

```python
#print the new dataframe with the encoded data to review
df_merged_marathon_air_weather_with_encoded
```

```python
#check the datatypes of the new dataframe
df_merged_marathon_air_weather_with_encoded.dtypes
```

```python
#print a correlation matrix / heatmap with the new encoded age groups
fig, ax = plt.subplots(figsize=(22,16))
sns.heatmap(df_merged_marathon_air_weather_with_encoded.corr(method='pearson'), annot=True, cmap='coolwarm');
```

# Strength of correlation
Regarding the strenght of the correlation coefficient pearson r, the following table can be used as a guide:
r	        Strength of correlation
0,0 < 0,1	no correlation
0,1 < 0,3	small correlation
0,3 < 0,5	medium correlation
0,5 < 0,7	high correlation
0,7 < 1	    very high correlation
Source: Kuckartz et al.: Statistik, Eine verständliche Einführung, 2013, p. 213
