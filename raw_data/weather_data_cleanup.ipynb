```python
import pandas as pd
import numpy as np
```

```python
weather2023 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 0)
weather2022 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 1)
weather2021 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 2)
weather2019 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 3)
weather2018 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 4)
weather2017 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 5)
weather2016 = pd.read_excel('./Weather data_v1.xlsx',sheet_name = 6)
```

```python
weather2023 = weather2023.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2022 = weather2022.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2021 = weather2021.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2019 = weather2019.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2018 = weather2018.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2017 = weather2017.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
weather2016 = weather2016.drop(labels = 1,axis = 0).drop(labels = ['Unnamed: 1'],axis = 1).reset_index()
```

```python
weather2023[['Date','Year']] = [pd.Timestamp('2023-04-23'),'2023']
weather2022[['Date','Year']] = [pd.Timestamp('2022-04-24'),'2022']
weather2021[['Date','Year']] = [pd.Timestamp('2021-09-12'),'2021']
weather2019[['Date','Year']] = [pd.Timestamp('2019-04-28'),'2019']
weather2018[['Date','Year']] = [pd.Timestamp('2018-04-29'),'2018']
weather2017[['Date','Year']] = [pd.Timestamp('2017-04-23'),'2017']
weather2016[['Date','Year']] = [pd.Timestamp('2016-04-17'),'2016']
```

```python
weather2023.head(5)
```

```python
all_weathers = pd.concat([weather2023,
                              weather2022,
                              weather2021,
                              weather2019,
                              weather2018,
                              weather2017,
                              weather2016],ignore_index = True)
```

```python
all_weathers.count()
```

```python
all_weathers.head()
```

```python
cols = ['Temp','Wind','Barometer']
all_weathers[cols] = all_weathers[cols].replace('\D','',regex = True)
all_weathers[cols] = all_weathers[cols].apply(pd.to_numeric)
```

```python
all_weathers.head()
```

```python
all_weathers.dtypes
```

```python
all_weathers['Date_Uhrzeit'] = all_weathers['Date'].astype(str) + ',' + all_weathers['Uhrzeit'].astype(str)
```

```python
all_weathers['Date_Uhrzeit'] = pd.to_datetime(all_weathers['Date_Uhrzeit'],format = "%Y-%m-%d,%H:%M:%S")
```

```python
all_weathers.head()
```

```python
all_weathers.rename(columns = {'Temp':'Temp_Celcius','Wind':'Wind_kmperhour','Barometer':'Barometer_hPa'},inplace = True)
```

```python
all_weathers.head()
```

```python
cols = ['Year','Date','Uhrzeit','Date_Uhrzeit','Wetter','Temp_Celcius','Wind_kmperhour','Luftfeuchte','Barometer_hPa']
all_weathers = all_weathers[cols]
```

```python
all_weathers.to_csv('weatherdata_2023_2016.csv',index = False)
```
